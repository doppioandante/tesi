```python
import numpy as np
import matplotlib.pyplot as plt

freq = 440
sampling = 10000
offset = 0.1

t_max = 30500/freq
ts = np.arange(0, t_max, 1/sampling)
y1 = np.sin(2*np.pi*freq*ts)
y2 = np.sin(2*np.pi*(freq+offset)*ts)
y1_energy = np.sum(y1**2)
err_energy = np.sum((y1-y2)**2)
plt.plot(ts, y1-y2)
plt.show()
print(y1_energy/err_energy)
```
