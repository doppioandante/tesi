```python
from scipy import signal
import numpy as np
from math import pi
import matplotlib.pyplot as plt
from scipy.fftpack import fft
```

```python
cutoff_frequency = 15000 # in Hertz
clock_frequency = 44000
b, a = signal.butter(4, cutoff_frequency/(clock_frequency/2), 'low')
w, h = signal.freqz(b, a, worN=2000)

fig, ax1 = plt.subplots()
ax1.set_title('Butterworth filter frequency response')
ax1.semilogx((clock_frequency * 0.5 / np.pi) * w, 20 * np.log10(abs(h)), 'b')
ax1.set_ylabel('Amplitude [dB]', color='b')
ax1.set_xlabel('Frequency [Hz]')
fig, ax2 = plt.subplots()
angles = np.unwrap(np.angle(h))
ax2.plot((clock_frequency * 0.5 / np.pi) * w, angles, 'g')
ax2.set_ylabel('Angle (radians)', color='g')
ax2.grid()
ax2.axis('tight')
plt.show()

```
